```python
import numpy as np
import pickle
import os
import time
import glob
import matplotlib.pyplot as plt
```

```python
logdir_prefix = 'lab-01'

## set your own path to data dir ##
#data_path ='/home/roblab12/Documents/repose/robotics_kinova_lab/Lab1/data' 
data_path = 'F:/DATA/robotics_lab/robotics_kinova_lab/Lab1/data'

data = []

for file in glob.glob(data_path + '/*/*.pkl'):
    with open(file, 'rb') as h:
        data.append(pickle.load(h))
```

```python
def plot_xyz(path_XYZ, num):

    fig = plt.figure(figsize = (45,15))
    ax1 = fig.add_subplot(2, 2, 1, projection='3d')
    ax1.scatter(path_XYZ[:,0],path_XYZ[:,1], path_XYZ[:,2], c='red', s=2) 
    ax1.plot(path_XYZ[:,0],path_XYZ[:,1], path_XYZ[:,2], color='black')
    ax1.set_xlabel('X[m]')
    ax1.set_ylabel('Y[m]')
    ax1.set_zlabel('Z[m]')
    ax1.set_title(f"XYZ Gripper path {num}")
    plt.show()
    
def plot_angles(path_ang, num):

    s_1 = np.linspace(0, len(path_ang[:,0]), num=len(path_ang[:,0])) # 3HZ sampling
    t_1 = s_1/3
    fig1 = plt.figure(figsize = (45,15))
    ax2 = fig1.add_subplot(4, 4, 1)
    ax2.plot(t_1, path_ang[:,0], color='black')
    ax2.plot(t_1, path_ang[:,1], color='blue')
    ax2.plot(t_1, path_ang[:,2], color='green')
    ax2.plot(t_1, path_ang[:,3], color='red')
    ax2.plot(t_1, path_ang[:,4], color='yellow')
    ax2.plot(t_1, path_ang[:,5], color='brown')
    ax2.set_xlabel('time[sec]')
    ax2.set_ylabel('Angle[deg]')
    ax2.set_title(f"Joints Angles path {num}")
    ax2.legend(['Joint_1','Joint_2','Joint_3','Joint_4','Joint_5','Joint_6'],loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()

def plot(path_xyz,coordinate):
    s_1 = np.linspace(0, len(path_xyz[:,coordinate]), num=len(path_xyz[:,coordinate])) # 3HZ sampling
    t_1 = s_1/3
    fig1 = plt.figure(figsize = (45,15))
    ax2 = fig1.add_subplot(2, 2, 1)
    ax2.plot(t_1, path_xyz[:,coordinate], color='black')
    
```

```python
for i in range(5):
    plot_xyz(data[i][1], i+1)
    plot_angles(data[i][0], i+1)
    plot(data[i][0])
    plot(data[i][1])
    plot(data[i][2])
```

```python
# work space

fig = plt.figure(figsize = (45,15))
ax3 = fig.add_subplot(2, 2, 1, projection='3d')
for j in range(5):
    ax3.scatter(data[j][1][:,0],data[j][1][:,1], data[j][1][:,2], c='black', s=2)
ax3.set_xlabel('X[m]')
ax3.set_ylabel('Y[m]')
ax3.set_zlabel('Z[m]')
ax3.set_title("XYZ Work Space of the gripper")

plt.show()
```
